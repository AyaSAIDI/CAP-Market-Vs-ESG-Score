```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Initialisation du DataFrame en dehors de la boucle
df = pd.DataFrame()

for i in range(1, 21):
    url = 'https://disfold.com/world/companies/?page=' + str(i)  # Concaténation correcte de 'i'
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')  # 'html.parser' est utilisé pour éviter un avertissement
    table = soup.find('table')
    world_titles = table.find_all('th')
    world_table_titles = [title.text.strip() for title in world_titles]

    column_data = table.find_all('tr')
    data = []
    for row in column_data[1:]:
        row_data = row.find_all('td')
        individual_row_data = [data.text.strip() for data in row_data]
        if len(individual_row_data) == len(world_table_titles):
            data.append(individual_row_data)
        else:
            print("Nombre de colonnes mismatched pour la ligne suivante, ligne ignorée :", individual_row_data)

    # Ajouter les données de cette page au DataFrame
    df = pd.concat([df, pd.DataFrame(data, columns=world_table_titles)], ignore_index=True)

# Exporter le DataFrame au format CSV
df.to_csv('TOP1000Companies.csv', index=False)
```
